```python
def fahrenheit (T):
    return((float(9)/(5)))*(T-275) +32.0
def celsius(T):
    return(float(5)/9)*((T-275)-32)

temperatures = [36.5, 37, 37.5, 38, 39]
F = list(map(fahrenheit, temperatures))
C = list (map(celsius, F))

print(F)
print(C)
```

```python
import numpy as np
from math import sin, cos, tan, pi
def map_function(x, functions):
    """map en iterable of functions on the object x"""
    res = []
    for i in x:
        for func in functions:
            res.append(func(1))
    return res
```

```python
values = np.linspace(0,pi,100)
family_functions = (sin, cos, tan)
print(map_function(values, family_functions))
```

## Reducing
import functools reduce
'reduce(func, seq)'

```python
from functools import reduce

reduce(lambda x, y: x+y, [47,11,42,43])
```

```python
num = [1,6,2323,45634,34524243,5475745]

reduce(lambda x,y: x if (x>y) else y, num)
```

```python
avengers = ["Captain America","Iron Man","Spiderman","Hulk","Thor"]
reduce(lambda x , y : x+y, avengers)
```
