```python
def mensaje(mensaje):
    MENSAJE = []
    for letter in mensaje:
        if letter.islower():
            MENSAJE.append(letter.upper())
        else: 
            MENSAJE.append(letter.lower())
    return "".join(MENSAJE)
```

```python
def shout(word1, word2):
    output1 = word1 + word2
    output2 = word2 + word1
    return output1, output2

```

```python
def factorial(number): #calcular un numero factorial con nestted functions
    if not isinstance(number, int): #s
        print("sos boludo, mete numeros enteros")
    if not number>=0:
        print("sos boludo los factoriales negativos no existen")
    def inner_factorial(number):
        if number <=1:
            return 1
        else:
            return number*inner_factorial(number-1)
    return inner_factorial(number)
```

```python
factorial(4)
```

```python
def outer():
    def inner():
        nonlocal x
        x = "superlocal"
        print("x es inner function", x)
    inner()
    x = x * 2
    print("x es outer fuction", x)
```

```python
outer()
```

# Default and  Flexible Parameters

```python
def saltoLista(N, salto = 1):
    lista = []
    for i in range(0,N, salto):
            lista.append(i)
    return lista
```

```python
print(saltoLista(10))
```

```python
def add_all(*args):
    sum_all = 0
    for i in args:
        sum_all +=i
    return sum_all
```

```python
add_all(1,2,3)
```

```python
def informacion(*args): 
    datos = {
    'nombre': args[0],
    'apellido': args[1],
    'edad': args[2],
    'genero': args[3],
 }
    return datos
```

```python
informacion("andres","ovalle",24,"M")
```

```python
def print_all(**kwargs):
    result = kwargs.get("string") * kwargs.get("N")
    if kwargs.get("reverse"):
        return result[::-1]
    else:
        return result
```

```python
print_all(string= "sergio", N = 3, reverse = True)
```
